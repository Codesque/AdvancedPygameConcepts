There is no module in python or in pygame that executes some code after some time lasts however we can build our own timers with some other tool in Pygame.

__IMPORTANT :__ We dont use <= or  >= sign because we measure the time in a discreate way which is fps . So that we probably end up with a number which is more likely to be bigger or smaller than the thresholder rather than being equal to . 

> __Static Point__ : is the point that we identfy after starting of the program

> __Current Time__ : Static Point + the amount of time that we want to measure

```python
import pygame 

pygame.init()  
screen = pygame.display.set_mode((600,600))
clock = pygame.time.Clock()   
running = True 
while running : 
    for e in pygame.event.get(): 
        if e.type == pygame.QUIT: 
            running = False 
            pygame.quit() 
            quit() 
    
    current_time = pygame.time.get_ticks() 
    print(current_time) 

    pygame.display.flip() 
    clock.tick(60)
```

> 1000/fps(60) = 17 , and the timer increases on every iteration by 17. 

```python
try :
    import pygame  
    import random 
    pygame.init() 
    screen = pygame.display.set_mode((500,500)) 
    bg_color = (0,0,0) 
    screen.fill(bg_color) 
    clock = pygame.time.Clock() 
    button_pressed_time = 0  
    current_time = 0 
    CHANGE_TIME = 2000  
    change_event = False 

    running = True 
    while running: 
        for e in pygame.event.get():
            if e.type == pygame.QUIT: 
                running = False 
                pygame.quit() 
                quit() 

            if e.type == pygame.KEYDOWN: 
                
                button_pressed_time = pygame.time.get_ticks()
                if not change_event: 
                    change_event = True    

            
        current_time = pygame.time.get_ticks()

        if change_event : 
                r , g  , b = bg_color 
                r += random.randint(-2,+2) 
                r = max(min(r , 255) , 0)   
                g += random.randint(-2,+2) 
                g = max(min(g, 255) , 0) 
                b += random.randint(-2,+2) 
                b = max(min(b , 255) , 0)  

                bg_color = r , g , b 

                screen.fill(bg_color) 

        if current_time - button_pressed_time > CHANGE_TIME : 
            
            change_event = False   

        pygame.display.update() 
        clock.tick(60)  

except : 
    pass 
```
